```python
# import wordclouds
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
import pandas as pd
# read data
X_train = pd.read_csv('Data/X_train.csv')
X_test = pd.read_csv('Data/X_test.csv')
y_train = pd.read_csv('Data/y_train.csv').rename(columns={'0': "Class_label"}, errors="raise")
y_test = pd.read_csv('Data/y_test.csv').rename(columns={'0': "Class_label"}, errors="raise")
```

```python
# merge all the datasets
train = pd.concat([X_train, y_train], axis=1)
test = pd.concat([X_test, y_test], axis=1)
df = pd.concat([train, test], axis=0)
df
```

```python
# seperate the dataset by their class
fake_df = df[df['Class_label'] == 1]
true_df = df[df['Class_label'] == 0]
true_df
```

```python
# drop the class and add all the frequency up
fake_df = fake_df.drop(['Class_label'], axis=1).sum(axis=0)
true_df = true_df.drop(['Class_label'], axis=1).sum(axis=0)
true_df
```

```python
# transpose the dataframe
Fake_news = fake_df.transpose()
True_news = true_df.transpose()
True_news.head()
```

```python
# produce wordcloud
wc = WordCloud(background_color="white", colormap="Dark2", max_font_size=150, random_state=42)

plt.rcParams['figure.figsize'] = [20, 15]

dfs = [True_news, Fake_news]
titles = ['True_news', 'Fake_news']

fig = plt.figure()
for i in range(2):
    wc.generate_from_frequencies(dfs[i])
    
    ax = fig.add_subplot(1,2,i+1)
    ax.imshow(wc, interpolation="bilinear")
    ax.axis("off")
    ax.set_title(titles[i])
    
fig.savefig('WordClouds.png')
```
